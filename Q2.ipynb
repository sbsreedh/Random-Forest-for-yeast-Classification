```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.metrics import mean
```

```python

train_accuracy=[]
test_accuracy=[]
mean_test_error = []
mean_train_error=[]
std_error=[]
std_dev=[]
test_scores = [] 
train_scores=[]
N = list(range(1, 31, 1))
for i in range(1,31):
    test_scores = [] 
    train_scores=[]
    test_std=[]
    for t in range(1,11):
        x_train_bag,x_test_bag,y_train_bag,y_test_bag=train_test_split(x_train,y_train,train_size=0.333,stratify=y_train)
        rf = RandomForestClassifier(n_estimators = i,max_features=3,bootstrap=True)
        rf.fit(x_train_bag, y_train_bag.values.ravel())
        predict_train=rf.predict(x_train_bag)
        predict_test=rf.predict(x_test)
       
        test_accuracy=accuracy_score(predict_test,y_test)
        train_accuracy=accuracy_score(predict_train,y_train_bag)
        
        
        test_std.append(1-test_accuracy)
        
        
        test_scores.append(metrics.accuracy_score(predict_test, y_test))
        train_scores.append(metrics.accuracy_score(predict_train, y_train_bag))
    #print(test_std)
   # print('STD:')
    #print(stdev(test_std))
    std_dev.append(stdev(test_std))
    #print(std_dev)
    test_error= [1 - t for t in test_scores]
    mean_test_error.append(mean(test_error))
    train_error= [1 - t for t in train_scores]
    mean_train_error.append(mean(train_error)) 
    std_error=statistics.stdev(test_std)
plt.plot(N,std_dev)
plt.xlabel('Number of trees')
plt.ylabel( 'Error rate std')
plt.show()
plt.plot(N, mean_train_error,label='Training Error')
plt.plot(N, mean_test_error,label='Testing Error')
plt.legend(loc='best')
plt.xlabel('Number of trees')
plt.ylabel('Mean error rate')
plt.show()       

        
```

We see that the error decreases as the the number of trees grow in the Forest!
